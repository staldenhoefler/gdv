```python
import pandas as pd
df = pd.read_csv('housing.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# get only the first 100 rows

small_df = df[:500]

# Streudiagramm mit unterschiedlichen GrÃ¶ÃŸen der Datenpunkte
plt.figure(figsize=(10, 6))
scatter = plt.scatter(small_df['median_house_value'], small_df['total_bedrooms'], s=small_df['median_income']*10, c='blue', alpha=0.2, edgecolors='w', linewidth=0.5)

# Titel und Achsenbeschriftungen
plt.title('Streudiagramm mit GrÃ¶ÃŸenunterschieden', fontsize=15)
plt.xlabel('Median House Value [USD]', fontsize=12)
plt.ylabel('Total Bedrooms', fontsize=12)

# Legende erstellen
handles, labels = scatter.legend_elements(prop="sizes", alpha=0.6)
legend1 = plt.legend(handles, labels, title="Median Income [USD]*10", loc="upper right")

# Diagramm anzeigen
plt.grid(True)
plt.show()
```

```python
# Make plot with different shapes. Use random datapoints.
import matplotlib.pyplot as plt
import numpy as np

# get 500 random rows

small_df = df.sample(n=500, random_state=0)

markers = {'NEAR BAY': 'o', '<1H OCEAN': 's', 'INLAND': '^', 'NEAR OCEAN': 'D', 'ISLAND': 'P'}
small_df['ocean_proximity'].unique()

plt.figure(figsize=(10, 6))
for cls in small_df['ocean_proximity'].unique():
    subset = small_df[small_df['ocean_proximity'] == cls]
    plt.scatter(subset['median_house_value'], subset['total_bedrooms'], label=f'Class {cls}', marker=markers[cls], alpha=0.8, edgecolors='w', linewidth=0.5)
    
# Titel und Achsenbeschriftungen
plt.title('Streudiagramm mit unterschiedlichen Formen', fontsize=15)
plt.xlabel('Median House Value [USD]', fontsize=12)
plt.ylabel('Total Bedrooms', fontsize=12)
plt.legend()

# Diagramm anzeigen
plt.grid(True)
plt.show()
```
