```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('housing.csv')
```

Barplot with binning

```python
df['median_income'].hist(bins=50)
plt.ylabel('Frequency')
plt.xlabel('Median Income')
plt.title('Histogram of Median Income Bins: 50')
plt.show()
```

```python
df['median_income'].hist(bins=200)
plt.ylabel('Frequency')
plt.xlabel('Median Income')
plt.title('Histogram of Median Income Bins: 200')
plt.show()
```

```python
df['median_income'].hist(bins=10)
plt.ylabel('Frequency')
plt.xlabel('Median Income')
plt.title('Histogram of Median Income Bins: 10')
plt.show()
```
