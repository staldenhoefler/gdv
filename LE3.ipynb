```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('housing.csv')
```

```python
df
```

Barplot with binning

```python
df['median_income'].hist(bins=50)
plt.ylabel('Frequency')
plt.xlabel('Median Income')
plt.title('Histogram of Median Income Bins: 50')
plt.show()
```

```python
df['median_income'].hist(bins=200)
plt.ylabel('Frequency')
plt.xlabel('Median Income')
plt.title('Histogram of Median Income Bins: 200')
plt.show()
```

```python
df['median_income'].hist(bins=10)
plt.ylabel('Frequency')
plt.xlabel('Median Income')
plt.title('Histogram of Median Income Bins: 10')
plt.show()
```

```python
# Make three bins: lower, middle, upper
categories = pd.cut(df['median_income'], bins=3, labels=['lower', 'middle', 'upper'])
df['income_category'] = categories

df['income_category'].value_counts().plot(kind='bar')
plt.ylabel('Frequency')
plt.xlabel('Income Category')
plt.title('Barplot of Income Category')
plt.show()
```

```python
import seaborn as sns
```

```python
# Add mean for median_house_value and median_income with line
mean_house_value = df['median_house_value'].mean()
mean_income = df['median_income'].mean()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='median_income', y='median_house_value', data=df)
plt.axvline(mean_income, color='red', linestyle='--', label='Mean Income')
plt.axhline(mean_house_value, color='blue', linestyle='--', label='Mean House Value')
plt.title('Scatterplot with Raw Data and Mean Lines')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.legend()
plt.show()
```

```python
# Scatterplot with raw data
plt.figure(figsize=(10, 6))
sns.scatterplot(x='median_income', y='median_house_value', data=df)
plt.title('Scatterplot with Raw Data')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()
```

```python
# Remove outliers
df_cleaned = df[df['median_house_value'] < 490000]
df_cleaned = df_cleaned[df_cleaned['median_income'] < 9]
df_cleaned = df_cleaned[df_cleaned['median_house_value'] > 45000]
df_cleaned = df_cleaned[df_cleaned['median_income'] > 0.5]
```

```python
# Scatterplot with cleaned data
plt.figure(figsize=(10, 6))
sns.scatterplot(x='median_income', y='median_house_value', data=df_cleaned)
plt.title('Scatterplot with Cleaned Data')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()
```

```python
# Add mean for median_house_value and median_income with line
mean_house_value = df_cleaned['median_house_value'].mean()
mean_income = df_cleaned['median_income'].mean()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='median_income', y='median_house_value', data=df_cleaned)
plt.axvline(mean_income, color='red', linestyle='--', label='Mean Income')
plt.axhline(mean_house_value, color='blue', linestyle='--', label='Mean House Value')
plt.title('Scatterplot with Cleaned Data and Mean Lines')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the scatter plot
plt.figure(figsize=(14, 7))
scatter_plot = sns.scatterplot(data=df, x='median_income', y='median_house_value',
                               hue='housing_median_age', style='ocean_proximity', palette='viridis')

plt.title('Scatter Plot of Median Income vs. Median House Value')
plt.xlabel('Median Income (in tens of thousands)')
plt.ylabel('Median House Value (in thousands)')
plt.legend(title='Housing Median Age')
plt.show()
```

```python
# Create the box plot
plt.figure(figsize=(14, 7))
box_plot = sns.boxplot(data=df, x='ocean_proximity', y='median_house_value', palette='pastel')

plt.title('Box Plot of Median House Value by Ocean Proximity')
plt.xlabel('Ocean Proximity')
plt.ylabel('Median House Value (in thousands)')
plt.show()
```

```python
import geopandas as gpd
from shapely.geometry import Point

# Convert to GeoDataFrame
geometry = [Point(xy) for xy in zip(df.longitude, df.latitude)]
geo_df = gpd.GeoDataFrame(df, geometry=geometry)

# Plot the data
plt.figure(figsize=(14, 7))
base = geo_df.plot(column='median_house_value', cmap='coolwarm', markersize=5, legend=True)
plt.title('Geospatial Plot of House Prices in California')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
